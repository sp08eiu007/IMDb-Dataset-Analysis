```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype
import warnings as w
w.filterwarnings("ignore")
```

```python
# read in 'imdb_top_1000.csv' and store it in a DataFrame named M
M=pd.read_csv("imdb_top_1000.csv")
M.head()
```

```python
# check the number of rows and columns
M.shape
```

```python
# check the data type of each column
M.dtypes
```

```python
# calculate the average movie rating
M['IMDB_Rating'].mean()
```

```python
# sort the DataFrame by runtime to find the shortest and longest movies
M.sort_values('Runtime')
```

```python
# create a histogram of IMDB_Rating, choosing an "appropriate" number of bins
M['IMDB_Rating'].plot(kind='hist', bins=10);
```

```python
# use a box plot to display that same data
M['IMDB_Rating'].plot(kind='box');
```

```python
# count how many movies have each of the IMDB ratings
M[['IMDB_Rating','Series_Title']].groupby('IMDB_Rating').count()
```

```python
# use a visualization to display that same data, including a title and x and y labels #fontsize=25
M[['IMDB_Rating','Series_Title']].groupby('IMDB_Rating').count().plot(kind='bar', title='IMDB Rating Visualization')
plt.xlabel('IMDB Rating')
plt.ylabel('Series Title Count');
```

```python
# calculate the average IMDB_Rating for each genre
M[['IMDB_Rating','Genre']].groupby('Genre').mean()
```

```python
# check if there are multiple movies with the same title, and if so, determine if they are actually duplicates
result = M[M['Series_Title'].isin(M[M.duplicated(['Series_Title'])]['Series_Title'])]
result.sort_values('Series_Title')
```
